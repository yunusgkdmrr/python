```python
import numpy as np


# Peaks function
def func_ex(x,y):
    f = 3*((1-x)**2)*np.exp(-x**2-(y+1)**2) - 10*(x/5-x**3-y**5)*np.exp(-x**2-y**2) - (1/3)*np.exp(-(x+1)**2-y**2)
    return f


# Discrete fourier transform 
def sigDFT(s):
    
    N = len(s)
    n = np.arange(0,N,1)
    S = np.zeros(N,dtype='complex128')
    
    # Calculate DFT
    for k in range(0,N):
        S[k] = np.sum( s*np.exp(-1j*2*np.pi*(k/N)*n) )
    
    return S

# Define the piecewise function
def pwisef(x):
    return np.piecewise(x, 
                        [(x < 1),  (x >= 1) & (x < 3),  (x >= 3) & (x <= 5)], 
                        [lambda x: x**2, lambda x: 3, lambda x: -5*x+8])

# Function using both *args and **kwargs together
def args_kwargs(q, p, *args, **kwargs):
    
    print("\nPositional arguments (args):")
    for arg in args:
        print(arg)

    print("\nKeyword arguments (kwargs):")
    for key, value in kwargs.items():
        print(f"{key} = {value}")
        
    return p**2, q**3




```

```python
# --- BURADAN SONRASINI KODUNUN ALTINA EKLE ---

# 1. func_ex testi
sonuc_peaks = func_ex(0.5, 0.5) 
print(f"Peaks Fonksiyonu Sonucu: {sonuc_peaks}")
print("-" * 30)

# 2. sigDFT testi (Basit bir sinyal uyduralım: [1, 2, 3, 4])
sinyal = np.array([1, 2, 3, 4])
dft_sonuc = sigDFT(sinyal)
print(f"DFT Sonucu:\n{dft_sonuc}")
print("-" * 30)

# 3. pwisef testi (Farklı aralıklardan sayılar verelim: 0.5, 2, 4)
x_degerleri = np.array([0.5, 2, 4])
piecewise_sonuc = pwisef(x_degerleri)
print(f"Parçalı Fonksiyon Sonucu: {piecewise_sonuc}")
# Açıklama: 0.5 -> karesi (0.25), 2 -> sabit (3), 4 -> -5*4+8 (-12)
print("-" * 30)

# 4. args_kwargs testi
print("Args ve Kwargs Testi:")
p_kare, q_kup = args_kwargs(2, 3, 10, 20, "ekstra bilgi", renk="kırmızı", yil=2025)
print(f"\nDönen değerler -> p^2: {p_kare}, q^3: {q_kup}")
```

```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt

# --- 1. Verileri Hazırla ---
xstart = 0
xstop = 2 * np.pi
increment = 0.1

x = np.arange(xstart, xstop, increment)
y = np.sin(x) # Gerçek Sinüs (Mavi)

# --- 2. Eğriyi Uydur (N=7 ile) ---
N = 3# Dereceyi 7 yaptık (Hata çok azalacak)
p = poly.polyfit(x, y, N)
y_tahmin = poly.polyval(x, p) # Polinomun tahmini (Turuncu)

# --- 3. Hatayı Hesapla (Kritik Nokta) ---
# Gerçek değerden tahmin edilen değeri çıkarıyoruz
hata = y - y_tahmin 

# --- 4. Çizim (Profesyonel Görünüm) ---
# figsize=(10, 8) diyerek ekranı dikeyde uzatıyoruz ki iki grafik sığsın
# 2 satır, 1 sütunluk bir yer açıyoruz (ax1: Üst, ax2: Alt)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# --- ÜST GRAFİK: Karşılaştırma ---
ax1.plot(x, y, label="Gerçek (Sinüs)", linewidth=2)
ax1.plot(x, y_tahmin, label=f"Tahmin (Polinom N={N})", linestyle="--", color="orange")
ax1.set_title("Ana Grafik: Gerçek vs Tahmin")
ax1.legend()
ax1.grid(True)

# İstersen üst grafikte çizgilerin ayrıldığını görmek için şunların başındaki # işaretini kaldır:
# ax1.set_xlim(1.4, 1.8)  # X ekseninde tepeye zoom yap
# ax1.set_ylim(0.99, 1.01) # Y ekseninde tepeye zoom yap


# --- ALT GRAFİK: Hata (Sapma) ---
# İşte sapmaları burada göreceksin
ax2.plot(x, hata, color="red", label="Sapma Miktarı (Residuals)")
ax2.set_title("Hata Analizi (Gerçek - Tahmin)")
ax2.set_xlabel("X Değerleri")
ax2.set_ylabel("Hata Büyüklüğü")
# Y ekseninde 0 çizgisini belirginleştirelim
ax2.axhline(0, color='black', linewidth=1, linestyle='--') 
ax2.legend()
ax2.grid(True)

# Grafikleri sıkışmasın diye aralarını aç
plt.tight_layout()
plt.show()
```

```python
import cmath

# 1. Karmaşık Sayı Oluşturma (Rectangular Form)
# z = x + jy (Python'da j kullanılır)
x = 3
y = 4
z = complex(x, y)

print(f"--- 1. Sayının Kendisi ---")
print(f"Sayı (z): {z}")
print(f"Reel Kısım (z.real): {z.real}")
print(f"Sanal Kısım (z.imag): {z.imag}")

# 2. Eşlenik (Conjugate)
# Sanal kısmın işaretini değiştirir (3 + 4j -> 3 - 4j)
print(f"\n--- 2. Eşlenik (Conjugate) ---")
print(f"Eşlenik: {z.conjugate()}")

# 3. Kutupsal Forma Çevirme (Polar Form)
# Bize (r, theta) yani (Genlik, Faz Açısı) verir.
# r = hipotenüs uzunluğu, theta = açı (radyan cinsinden)
r, theta = cmath.polar(z)

print(f"\n--- 3. Kutupsal Form (Polar) ---")
print(f"Genlik (r): {r}")        # 3-4-5 üçgeninden 5 gelir
print(f"Faz Açısı (theta): {theta} radyan")

# 4. Geri Çevirme (Rectangular Form)
# Elimizdeki Genlik ve Açı'dan tekrar x+jy elde etme
z_yeni = cmath.rect(r, theta)

print(f"\n--- 4. Geri Dönüş (Rect) ---")
print(f"Eski haline döndü: {z_yeni}")
```

```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt

# 1. Dosyayı Yazma Modunda Aç ("w")
# Eğer dosya yoksa oluşturur, varsa içini silip baştan yazar.
f = open("simdata.txt", "w")

# Dosyaya yazma fonksiyonu (Sekreter)
def writedata(t, x):
    time_str = str(t)
    value_str = str(round(x, 2)) # Virgülden sonra 2 basamak yuvarla
    # Araya TAB boşluğu (\t) koy, sonuna enter (\n) bas
    f.write(time_str + "\t" + value_str + "\n")

# 2. Model Parametreleri
T = 5
a = -1/T
x0 = 1

# 3. Simülasyon Ayarları
tstart = 0
tstop = 25
increment = 1

# Dizileri oluştur (NumPy ile)
# tstop+1 diyoruz çünkü 0'dan 25'e kadar (25 dahil) 26 sayı var.
t = np.arange(tstart, tstop + 1, increment)
x = np.zeros(len(t)) 

# 4. Hesaplama Döngüsü
# Slaytta range(tstop) demiş ama o zaman son saniyeyi hesaplamaz.
# Doğrusu len(t) kadar dönmektir.
for k in range(len(t)):
    # Formül: x(t) = x0 * e^(-t/T)
    x[k] = mt.exp(a * t[k]) * x0
    
    # Hesapladığını hemen dosyaya yaz
    writedata(t[k], x[k])

# İşimiz bitince dosyayı kapatıyoruz (Çok önemli!)
f.close()

# 5. Grafiği Çizdir
plt.plot(t, x)
plt.title('Simulation of Dynamic System')
plt.xlabel('t (Zaman)')
plt.ylabel('x (Değer)')
plt.grid(True)
plt.axis([0, 25, 0, 1]) # Eksen sınırları
plt.show()
```
